```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer_df =pd.DataFrame(  np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'],'target'))
```

```python
cancer_df.head(5)
```

```python
sns.pairplot(cancer_df,hue = 'target',vars =['mean radius','mean texture','mean area','mean smoothness','mean concavity'])
```

```python
sns.countplot(cancer_df['target'],label = 'count')
```

```python
sns.scatterplot(x = 'mean area',y = 'mean smoothness',hue = 'target', data = cancer_df)
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(cancer_df.corr(),annot = True)
```

```python
#model training
x =cancer_df.drop(['target'],axis = 1)
y = cancer_df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25)
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(x_train,y_train)
```

```python
#evaluate the data
y_predict = svc_model.predict(x_test)
```

```python
y_predict
```

```python
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test,y_predict))
```

```python
#improving the model.....we do this by normalizing data

min_train = x_train.min()
min_train
```

```python
range_train = (x_train - min_train).max()
```

```python
range_train
```

```python
x_train_scaled = (x_train - min_train)/range_train
```

```python
x_train_scaled
```

```python
#teh original data
sns.scatterplot(x = x_train['mean area'],y = x_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = x_train_scaled['mean area'],y = x_train_scaled['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = x_train['mean area'],y = x_train['mean smoothness'], hue = y_train)
```

```python
min_test = x_test.min()
range_test = (x_test - min_test).max()
x_test_scaled = (x_test - min_test)/range_test
```

```python
x_test_scaled 
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model = SVC()
svc_model.fit(x_train_scaled,y_train)
```

```python
y_predict = svc_model.predict(x_test_scaled)
cm = confusion_matrix(y_test,y_predict)

sns.heatmap(cm, annot = True, fmt ='d')
```

```python
print(classification_report(y_test,y_predict))
```
